```python
etnia=pd.crosstab(dados['paciente_racaCor_valor'], dados['vacina_descricao_dose'])



etnia.plot.bar(stacked=True)
plt.ylabel('Frequência absoluta em milhoes')
plt.xticks(rotation = 'horizontal')
plt.legend(loc='center')
plt.legend(bbox_to_anchor=(1.05, 0.95)); # Coloca legenda para fora da janela gráfica
plt.show()
```

Por serem muitos grupo, a representação grafica nao ficou tao legal, por isso peço que siga o crosstab e os dados numericos para perceber o foco o publico dos atuantes na area da saude e nos idosos


```python
lis=[]
for i in dados['vacina_descricao_dose']:
    if i not in lis:
        lis.append(i)
print(lis)
print(len(lis))

primeira_dose=dados.loc[dados['vacina_descricao_dose']==1.0]
segunda_dose=dados.loc[dados['vacina_descricao_dose']==2.0]

plt.figure(figsize=(15, 2*len(lis)))

i=0
while i < 8:
    num=int(f'81{i+1}')
    graf=dados.loc[dados['vacina_descricao_dose']==lis[i]]
    
    plt.subplot(num) # Janela gráfica com 2 linhas, 2 colunas e histograma a seguir é colocado na posição 1
    plt.hist(graf['paciente_idade'], bins=[0,15,30,45,60,75,90],edgecolor='white')
    plt.title(f"{lis[i]}")
    i+=1
plt.show()

plt.figure(figsize=(15, 2*len(lis)))

while i < 16:
    num=int(f'81{i-7}')
    graf=dados.loc[dados['vacina_descricao_dose']==lis[i]]
    
    plt.subplot(num) # Janela gráfica com 2 linhas, 2 colunas e histograma a seguir é colocado na posição 1
    plt.hist(graf['paciente_idade'], bins=[0,15,30,45,60,75,90],edgecolor='white')
    plt.title(f"{lis[i]}")
    i+=1
plt.show()

plt.figure(figsize=(15, 2*len(lis)))

while i < len(lis):
    num=int(f'{len(lis)-16}1{i-15}')
    graf=dados.loc[dados['vacina_descricao_dose']==lis[i]]
    
    plt.subplot(num) # Janela gráfica com 2 linhas, 2 colunas e histograma a seguir é colocado na posição 1
    plt.hist(graf['paciente_idade'], bins=[0,15,30,45,60,75,90],edgecolor='white')
    plt.title(f"{lis[i]}")
    i+=1
plt.show()

# plt.subplot(211) # Janela gráfica com 2 linhas, 2 colunas e histograma a seguir é colocado na posição 1
# plt.hist(primeira_dose['paciente_idade'], bins=[0,15,30,45,60,75,90],edgecolor='white')
# plt.title("Histograma 1")

# plt.subplot(212) # Janela gráfica com 2 linhas, 2 colunas e histograma a seguir é colocado na posição 1
# plt.hist(segunda_dose['paciente_idade'],bins=[0,15,30,45,60,75,90], edgecolor='white')
# plt.title("Histograma 2")

```
